# Object Detection

```python
from ultralytics import YOLO
import cv2

#Load the YOLOv8 model
model = YOLO('yolov8n.pt')
```

## Prediction(1)

```python
from PIL import Image
# from PIL
im1 = Image.open(r"C:\deep\elephant.jpg")
results = model.predict(source=im1, save=True)  # save plotted images
```

## Plot Predicted image (1)

```python
#print(results)
display(Image.open('runs/detect/predict/elephant.jpg'))
```
